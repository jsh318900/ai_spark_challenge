```python
root = '/content/drive/MyDrive/Colab Notebooks/AF_pj/dataset/'
label = ['공주','노은동','논산','대천2동','독곶리','동문동','모총동','문창동','성성동','신방동','신흥동','아름동','예산군','읍내동','이원면','정림동','홍성읍']
label_aws = ['계룡','공주','논산','당진','대산','대천항','북격렬비도','서천','성거','세종고운','세종연서','세종전의','세천','아산','안도','양화','예산','오월도','옹도','유구','장동','정산','정안','청양','춘장대','태안','호도','총북','홍성죽도']
```

```python
awsmap = pd.read_csv(root+'META/awsmap.csv')
pmmap = pd.read_csv(root+'META/pmmap.csv')
display(awsmap.head(3))
display(pmmap.head(3))
```

```python
train_dust_fp = glob.glob(root+'TRAIN/*')
train_aws_fp = glob.glob(root+'TRAIN_AWS/*')
train_dust, train_aws = {}, {}
for fp in train_dust_fp:
    k = fp.split('/')[-1][:-4]
    train_dust[k] = pd.read_csv(fp)
    train_dust[k]['Latitude'] = pmmap.loc[pmmap['Location']==k, 'Latitude'].item()
    train_dust[k]['Longitude'] = pmmap.loc[pmmap['Location']==k, 'Longitude'].item()
for fp in train_aws_fp:
    k = fp.split('/')[-1][:-4]
    train_aws[k] = pd.read_csv(fp)
    train_aws[k]['Latitude'] = awsmap.loc[awsmap['Location']==k, 'Latitude'].item()
    train_aws[k]['Longitude'] = awsmap.loc[awsmap['Location']==k, 'Longitude'].item()
len(train_dust), len(train_aws)
```

```python
train_dust['공주'].info()
```

```python
mydic_dust = {}
idx_dust = ['데이터 총계','pm결측값']
for k, df in train_dust.items():
    mydic_dust[k] = []
    mydic_dust[k].append(df['연도'].value_counts().sum())
    mydic_dust[k].append(df['PM2.5'].isna().sum())
    
pd.DataFrame(data=mydic_dust, index=idx_dust)
```

```python
train_aws['계룡'].info()
```

```python
mydic_aws = {}
idx_aws = ['데이터 총계','기온결측값','풍향결측','풍속결측','강수량결측','습도결측']
for k, df in train_aws.items():
    mydic_aws[k] = []
    mydic_aws[k].append(df['연도'].value_counts().sum())
    mydic_aws[k].append(df['기온(°C)'].isna().sum())
    mydic_aws[k].append(df['풍향(deg)'].isna().sum())
    mydic_aws[k].append(df['풍속(m/s)'].isna().sum())
    mydic_aws[k].append(df['강수량(mm)'].isna().sum())
    mydic_aws[k].append(df['습도(%)'].isna().sum())
    
pd.DataFrame(data=mydic_aws, index=idx_aws)
```

```python
ex = train_dust['공주'].copy()
ex['날짜'] = ex['일시'].apply(lambda x: x.split()[0])
ex['시간'] = ex['일시'].apply(lambda x: x.split()[1])
ex.head(3)
```
